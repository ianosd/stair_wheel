```python
import shape_computation
import plot
import matplotlib.pyplot as plt
import numpy

%matplotlib inline
```

Suppose we are dealing with some stairs, that have some particular horizontal and vertical dimensions:

```python
stair_width = 100
stair_height = 50

plot.plot_stairs(stair_width=stair_width, stair_height=stair_height, num_climbs=2)
```

We will search for shapes parameterized by an array of radii, $r$, spanned over an angular region of size $\theta$, which is repeated for $n$ times inside a $2\pi$ region. Let's name it a star wheel.
Here's an example:

```python
shape = shape_computation.StarWheel(n=3, theta=2*numpy.pi/3, r=1.1**numpy.linspace(0, 10, 11))
plot.plot_star_wheel(shape)
_ = plt.axis('equal')
```

Given particular stair dimensions, a fixed number of corners $n$ of our star wheel, we will look for the proper radii $r$ and angular opening $\theta$, such that the shape will roll as nicely as possible along the stairs. Note that the array of radii has fixed length.

```python
n = 5
number_of_radii = 10
computed_wheel = shape_computation.compute_star_wheel(n=n, number_of_radii=number_of_radii, stair_width=stair_width, stair_height=stair_height)
plot.plot_star_wheel(computed_wheel)
_ = plt.axis('equal')
```

The compute_star_wheel method is solving an optimisation problem whith constraints.

The optimised objective is the length of the shape's centre trajectory, as it rolls along the stairs.
The constraints ensure that the shape is rolling in sync with the stairs, and that the shape is convex

```python
from matplotlib import animation, rc
from IPython.display import HTML

animation = plot.create_animation(computed_wheel, stair_width, stair_height, n)
HTML(animation.to_jshtml())
```
